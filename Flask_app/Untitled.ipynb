```python
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy.orm import Session
```

```python
marvel_characters = pd.read_csv("../comic-characters_marvel-wikia-data.csv")
```

```python
marvel_characters.reset_index(drop=True)
marvel_characters.head()
```

```python
username = "postgres"
password = "password"
server = "marvelcharacters.cnzcqo7polqc.us-east-2.rds.amazonaws.com"
database = "marvelcharacters"
# create engine
engine = create_engine(f'postgresql+psycopg2://{username}:{password}@{server}/{database}')
# Create our session (link) from Python to the DB
session = Session(engine)
table_name = "marvel_characters"
marvel_characters.to_sql(table_name, con=engine, if_exists='append')
# Close Session
session.close()
```
